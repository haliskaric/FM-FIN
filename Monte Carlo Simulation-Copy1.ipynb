```python
import numpy as np 
import pandas as pd
from datetime import date

import random

import holoviews as hv

import yfinance as yf

```

```python
stocks = ['DIS', 'NFLX', 'PARA', 'CMCSA', 'ROKU', 'FOXA', 'DISCA', 'WMG', 'CURI', 'LYV', 'T']
```

```python
begin_date = '2021-01-01'
end_date = date.today()
```

```python
def generate_stock_returns(stocks, begin_date, end_date):
    
    prices = pd.DataFrame()
    
    for stock in stocks:
        df = yf.download(stocks, begin_date, end_date)
        df1 = df['Adj Close']
        return df1

df1 = generate_stock_returns(stocks, begin_date, end_date)
```

```python
df1
```

```python
returns = df1.pct_change()
returns
```

```python
cov = returns.cov()
cov.head()
```

```python
np.random.seed(10)
weights = np.random.random(len(stocks))
weights /= np.sum(weights)
weights
```

```python
rp = (returns.mean()*252)@weights
rp
```

```python
port_var = weights@(cov*252)@weights
port_var
```

```python
rf = 0.0185
sharpe = (rp-rf)/np.sqrt(port_var)
sharpe
```

```python
len(weights)
```

```python
def portfolio_metrics(weights, index='Trial'):
    '''
    Function which generates performance metrics and will be reported and be used to find optimale weights.
    
    '''
    rp = (returns.mean()*252)@weights
    port_var = weights@(cov*252)@weights
    sharpe = (rp-rf)/np.sqrt(port_var)
    df2 = pd.DataFrame({'Expected Return': rp,
                        'Portfolio Variance': port_var,
                        'Portfolio Std': np.sqrt(port_var),
                        'Sharpe Ratio': sharpe}, index=[index])
    return df2

portfolio_metrics(weights)
```

```python
np.random.seed(42)
portfolios = pd.DataFrame(columns=[*stocks, 'Expected Return', 'Portfolio Variance', 'Portfolio Std', 'Sharpe Ratio'])

for i in range(100000):
    weights = np.random.random(len(stocks))
    weights /= np.sum(weights)
    portfolios.loc[i, stocks] = weights
    metrics = portfolio_metrics(weights,i)
    portfolios.loc[i, ['Expected Return', 'Portfolio Variance', 'Portfolio Std', 'Sharpe Ratio']] = \
    metrics.loc[i, ['Expected Return', 'Portfolio Variance', 'Portfolio Std', 'Sharpe Ratio']]
    
portfolios
```

```python
optimal_portfolio = portfolios[portfolios['Sharpe Ratio']==portfolios['Sharpe Ratio'].max()].T
```

```python
optimal_portfolio
```

```python
opt_weights = portfolios[portfolios['Sharpe Ratio']==portfolios['Sharpe Ratio'].max()].to_numpy()[0][0:11]
```

```python
def optimal_portfolio_weights(weights):
    df = pd.DataFrame({'Stocks': stocks,
                       'Weights': weights})
    df.Weights = df.Weights.map(lambda x: '{:.2%}'.format(x))
    return df

optimal_portfolio_weights(opt_weights)
```
